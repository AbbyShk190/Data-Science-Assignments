```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
df = pd.read_csv("D:\\Data Sets\\Online retail.csv")
df
```

```python
# Split the single column into lists of items
df['Items'] = df.iloc[:, 0].str.split(',')

# Convert the dataset into a one-hot encoded format for association rules
te = TransactionEncoder()
te_ary = te.fit(df['Items']).transform(df['Items'])
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Apply the Apriori algorithm to find frequent itemsets with minimum support of 0.03
frequent_itemsets = apriori(df_encoded, min_support=0.03, use_colnames=True)

# Generate association rules with minimum confidence of 0.6
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.6, num_itemsets=len(frequent_itemsets))

# Filter and display meaningful rules (lift > 1.2, confidence > 0.6)
meaningful_rules = rules[(rules['lift'] > 1.2) & (rules['confidence'] > 0.6)]

# Display the frequent itemsets and meaningful association rules
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nMeaningful Association Rules:")
print(meaningful_rules)
```

```python
# Show the most interesting rules based on lift, confidence, and support
top_rules = meaningful_rules.sort_values(by='lift', ascending=False)

# Show top 10 rules
print("Top 10 Association Rules (sorted by Lift):")
print(top_rules.head(10))

# Further analysis based on confidence and support
# Filter rules with a very high confidence (>0.8) to find the strongest relationships
strong_rules = meaningful_rules[meaningful_rules['confidence'] > 0.8]

# Display strong association rules
print("\nStrong Association Rules (Confidence > 0.8):")
print(strong_rules)

# Analyze high support rules (> 0.05)
high_support_rules = meaningful_rules[meaningful_rules['support'] > 0.05]

# Display high support association rules
print("\nHigh Support Association Rules (Support > 0.05):")
print(high_support_rules)
```
